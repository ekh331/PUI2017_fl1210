```python
from __future__ import print_function
import numpy as np
import pandas as pd

import sys
import os
import pylab as pl
%pylab inline
```

```python
# check the environmental variable DFDATA
dfdata = os.getenv("DFDATA")
print (dfdata)
```

```python
#  read in the CSV file from the DF into a pandas dataframe
df = pd.read_csv(os.getenv("DFDATA") + "/y43c-5n92/1488844800/y43c-5n92.csv", parse_dates=['Date'], infer_datetime_format=True)
```

```python
df.head()
```

```python
# drop variables that do not need
df.drop(['Site', 'Turbidity(NTU) at 12AM', 'Turbidity(NTU) at 4AM', 'Turbidity(NTU) at 8AM', 'Turbidity(NTU) at 12PM', 'Turbidity(NTU) at 4PM',
         'Turbidity(NTU) at 8PM', 'Coliform, Fecal(fc/100mL)'], 
        axis=1, inplace=True)
```

```python
df.head()
```

```python
# plot average 24hr turbidity changes of water from 2015 to 2017
df.plot(x='Date', y='Average 24hrTurbidity(NTU)',figsize=(8,4), rot=90)
plt.title('Average 24 hr water turbidity 2015-2017',fontsize=12)
plt.xlabel('Date',fontsize=12)
plt.ylabel('Average 24hr Turbidity(NTU)',fontsize=12)
```
